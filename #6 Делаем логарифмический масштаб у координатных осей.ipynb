```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()

x = np.arange(-10 * np.pi, 10 * np.pi, 0.5)
y = np.sinc(x) * np.exp(-np.abs(x/10))

ax.plot(x, y)

ax.grid()
plt.show()
```
