## Making a logarithmic scale for coordinate axes_Theory

---

* **ax.semilogy(x, y)** – **y oxunu logaritmik miqyasda**, **x oxunu isə adi (linear) miqyasda** göstərən qrafik yaradır.

- Bu metod **eksponensial artım** və ya **çox böyük/müxtəlif diapazonlu dəyərləri** vizuallaşdırmaq üçün istifadə olunur.
- `y` dəyərləri **müsbət olmalıdır**, çünki logaritmik miqyasda sıfır və mənfi ədədlər istifadə edilə bilməz.

> Nəticədə, log-miqqyaslı y oxu olan qrafik alınır.

---

* **ax.semilogx(x, y)** – **x oxunu logaritmik**, **y oxunu isə adi (linear)** miqyasda göstərən qrafik yaradır.

- Bu metod **eksponensial şəkildə dəyişən x dəyərləri** üçün uyğundur (məsələn, 10, 100, 1000 və s.).
- `x` dəyərləri **müsbət olmalıdır**, çünki logaritmik miqyasda sıfır və mənfi ədədlər qəbul edilmir.

> Nəticədə, log-miqqaslı x oxu olan qrafik əldə olunur.

---

* **ax.plot(x, y)** – **x və y dəyərlərinə əsasən xətti qrafik** çəkir.

* **ax.set\_yscale('log')** – **y oxunu logaritmik miqyasda** göstərir.

  * Bu, `ax.semilogy(x, y)` ilə **eyni nəticəni verir**, amma `set_yscale()` metodu daha **fleksibldir** (məsələn, `'linear'`, `'log'`, `'symlog'`, `'logit'` və s. kimi fərqli miqyaslar üçün istifadə oluna bilər).

> Yəni əvvəlcə adi qrafik çəkilir, sonra isə y oxu logaritmik miqyasla dəyişdirilir.

---

* **ax.plot(x, y)** – `x` və `y` dəyərlərinə əsaslanaraq **xətti qrafik** çəkir.

* **ax.set\_xscale('log')** – **x oxunu logaritmik miqyasda** göstərir.

  * Bu metod `ax.semilogx(x, y)` ilə **eyni məna daşıyır**, lakin `set_xscale()` daha **ümumi və çevik** yanaşmadır.

> Beləliklə, qrafikdə **x oxu logaritmik**, **y oxu isə linear** olur. `x` dəyərləri **müsbət olmalıdır**, əks halda xəta verir.

---

* **ax.set\_yscale('log', base=10, subs=\[2, 9\])** – **y oxunu 10 əsaslı logaritmik miqyasda** göstərir və **aralıq (secondary) log xəttlərini** 2 və 9 üçün əlavə edir.

- `base=10` → log miqyası **10 əsasında** qurulur (məs., 10¹, 10², 10³...).
- `subs=[2, 9]` → hər log ədədi arasında **əlavə köməkçi xətlər** 2 və 9 üçün yerləşdirilir.
  Məsələn: 20, 90, 200, 900, 2000 və s.

> Bu, log miqyasda **daha ətraflı vizual bölünmələr** göstərmək üçün istifadə olunur.

---

* **ax.set\_xscale('symlog', linthresh=2)** – **x oxunu simmetrik logaritmik (symlog) miqyasda** göstərir.

  * `symlog` həm **müsbət**, həm də **mənfi** dəyərlərlə işləyə bilir, fərqli olaraq `log` miqyası yalnız müsbət dəyərləri qəbul edir.
  * `linthresh=2` → **-2 ilə +2 arasında** ox **linear miqyasda**,
    digər hissələr isə **logaritmik miqyasda** göstərilir.

> Bu, həm mənfi, həm sıfıra yaxın, həm də böyük dəyərlərin **bir qrafikdə səliqəli əks olunması** üçün istifadə olunur.

---

* **ax.set\_xscale('symlog', linthresh=2, linscale=5)** – **x oxunu simmetrik logaritmik miqyasda** göstərir və aşağıdakı tənzimləmələri edir:

- `linthresh=2` → **-2 ilə +2** arasındakı sahə **linear miqyasda** göstərilir.
- `linscale=5` → **linear hissənin miqyası log hissəyə nisbətən 5 dəfə daha geniş** göstərilir.
  Yəni, keçid zonası daha **"yumşaq"** olur.

> Bu parametr **sıfıra yaxın və həm müsbət, həm mənfi dəyərlərin eyni qrafikdə əksini daha vizual rahat** görmək üçün istifadə olunur.

---

* **ax.loglog(x, y)** – Həm **x**, həm də **y** oxlarını **logaritmik miqyasda** göstərən qrafik yaradır.

  * Bu metod həm `ax.set_xscale('log')`, həm də `ax.set_yscale('log')` funksiyalarını eyni anda tətbiq edir.

> **Eksponensial və ya potensial funksiyaların** davranışını təhlil etmək və **ölçüləri geniş fərqlənən** dəyərləri vizuallaşdırmaq üçün istifadə olunur.

> `x` və `y` dəyərləri **müsbət olmalıdır**, əks halda xəta verir.
