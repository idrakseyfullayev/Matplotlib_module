## Boundary values __of axes and locators_Theory

---

* **ax.set(xlim=(-5, 30), ylim=(-1, 6))** – `ax` obyektində x oxunun sərhədlərini -5-dən 30-a, y oxunun sərhədlərini isə -1-dən 6-ya qədər təyin edir.
- `ax.set(xlim=(-5, 30), ylim=(-1, 6))`

---

* **ax.set\_xlim(-5, 30)** – x oxunun minimum və maksimum sərhədlərini -5 və 30 olaraq təyin edir.
- `ax.set_xlim(-5, 30)`

* **ax.set\_ylim(-1, 6)** – y oxunun minimum və maksimum sərhədlərini -1 və 6 olaraq təyin edir.
- `ax.set_ylim(-1, 6)`

---

* **plt.xlim(-5, 20)** – Qrafikdə x oxunun görünən sərhədlərini -5-dən 20-yə qədər təyin edir.
  `plt.xlim(-5, 20)`

* **plt.ylim(-1, 6)** – Qrafikdə y oxunun görünən sərhədlərini -1-dən 6-ya qədər təyin edir.
  `plt.ylim(-1, 6)`

---

* **ax.set\_xlim(xmin=-5, xmax=10)** – x oxunun minimum sərhədini -5, maksimum sərhədini isə 10 olaraq təyin edir.
- `ax.set_xlim(xmin=-5, xmax=10)`

---

* **lc = NullLocator()** – Heç bir əsas (major) tix (ox nişanı) göstərməmək üçün `NullLocator` obyektini yaradır.
- `lc = NullLocator()`

* **ax.xaxis.set\_major\_locator(lc)** – x oxundakı bütün əsas nişanları (major ticks) gizlədir.
- `ax.xaxis.set_major_locator(lc)`

---

* **ax.yaxis.set\_major\_locator(LinearLocator(5))** – y oxunda **bərabər aralıqlı** olmaqla **5 əsas nişan** (major tick) göstərir.
- `LinearLocator(numticks)` – burada:

  * `numticks`: göstəriləcək əsas nişanların (ticks) sayı.

---

* **MultipleLocator(base=1)** – Oxa **müəyyən aralıqla** (`base`) əsas nişanlar yerləşdirir.
- `ax.xaxis.set_major_locator(MultipleLocator(base=1))` – x oxunda hər 1 vahiddən bir əsas nişan göstərir.

---

* **IndexLocator(base=0.5, offset=0)** – Oxa **sabit aralıqla** (`base`) və **müəyyən başlanğıc nöqtəsindən** (`offset`) başlayaraq əsas nişanlar yerləşdirir.
- `ax.xaxis.set_major_locator(IndexLocator(base=0.5, offset=0))` – x oxunda 0-dan başlayaraq hər 0.5 vahiddən bir əsas nişan göstərir.

---

* **FixedLocator(locs)** – Oxa yalnız **verilən koordinatlarda** əsas nişanlar (major ticks) yerləşdirir.
- `ax.xaxis.set_major_locator(FixedLocator([-2, -1, 0, 1, 2, 3]))`

  **Parametrlər:**

  * `locs`: əsas nişanların göstəriləcəyi **dəqiq koordinat dəyərlərinin siyahısıdır**.

---

* **LogLocator(base=2)** – Oxa **logaritmik ölçüdə**, verilmiş **əsasa** (`base`) uyğun əsas nişanlar yerləşdirir.
- `ax.xaxis.set_major_locator(LogLocator(base=2))` – x oxunda **2 əsaslı logaritmik aralıqlarla** əsas nişanlar göstərir.

  **Parametrlər:**

  * `base`: logaritmin əsası (məsələn, 2, 10 və s.)
  * Digər isteğe bağlı parametrlər: `subs`, `numticks`, `labelOnlyBase`

---

* **MaxNLocator(5)** – Oxda **ən çox** 5 əsas nişan (major tick) göstərməyə çalışır, avtomatik olaraq dəyərləri seçir.
- `ax.xaxis.set_major_locator(MaxNLocator(5))`

  **Parametrlər:**

  * `nbins=5` – göstəriləcək maksimum əsas nişanların sayı
  * `steps`, `integer`, `prune`, `symmetric` kimi əlavə parametrlərlə də tənzimlənə bilər.

---

* **ax.minorticks\_on()** – Həm x, həm də y oxlarında **minor (kiçik)** nişanları (ticks) aktivləşdirir.
- `ax.minorticks_on()`

* **ax.grid(which='major', lw=2)** – Yalnız **əsas (major)** xətlər üçün şəbəkə (grid) göstərir və qalınlığını `2` təyin edir.
- `ax.grid(which='major', lw=2)`

* **ax.grid(which='minor')** – Yalnız **kiçik (minor)** xətlər üçün şəbəkə göstərir.
- `ax.grid(which='minor')`

* **ax.xaxis.set\_major\_locator(MaxNLocator(7))** – x oxunda maksimum **7 əsas nişan** göstərir.
- `MaxNLocator(7)` – avtomatik olaraq ən uyğun 7 dəyəri seçir.
