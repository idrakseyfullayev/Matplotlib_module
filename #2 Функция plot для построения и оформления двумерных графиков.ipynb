```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
y = np.cos(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
# plt.fill_between(x, y)
plt.grid()
plt.show()
```

```python
plt.plot(x, y)
plt.fill_between(x, y)
plt.grid()
plt.show()
```

```python
plt.plot(x, y)
plt.fill_between(x, y, color = 'r')
plt.grid()
plt.show()
```

```python
plt.plot(x, y)
plt.fill_between(x, y, color = 'r',  0.5)
plt.grid()
plt.show()
```

```python
plt.plot(x, y)
plt.fill_between(x, y, 0.5, color = 'r')
plt.grid()
plt.show()
```

```python
plt.plot(x, y)
plt.fill_between(x, y, 0.5, where= y < 0, color = 'r')
plt.grid()
plt.show()
```

```python
plt.plot(x, y)
plt.fill_between(x, y, where= y < 0, color = 'r')
plt.grid()
plt.show()
```

```python
plt.plot(x, y)
plt.fill_between(x, y, where= y < 0, color = 'r', alpha = 0.5)
plt.fill_between(x, y, where= y > 0, color = 'g', alpha = 0.5)
plt.grid()
plt.show()
```
