## Add a legend_Theory

* **ax.legend()** – Qrafikdəki legend göstərmək üçün istifadə olunur.

- `ax.legend()`
- Burada:

  * Qrafikdəki bütün etiketlənmiş xətt və ya obyektlərin adları avtomatik olaraq legend şəklində əlavə edilir və görünür.

---

* **ax.plot()** – Verilmiş verilənləri qrafik üzərində xətt kimi çəkmək üçün istifadə olunur.

- `ax1 = ax.plot(np.arange(0, 5, 0.25), label='line1')`
- `ax2 = ax.plot(np.arange(0, 10, 0.5), label='line2')`
- Burada:

  * `np.arange(0, 5, 0.25)` və `np.arange(0, 10, 0.5)` — xətlərin çəkilməsi üçün istifadə olunan nöqtələr
  * `label='line1'` və `label='line2'` — hər xətt üçün legend-də göstəriləcək adlar
  * `ax1`, `ax2` — çəkilmiş xətt obyektlərinin siyahısı (list) kimi qaytarılır.

---

* **ax.legend()** – Qrafikdə legend göstərmək üçün istifadə olunur və əfsanədə göstəriləcək etiketləri əl ilə təyin etmək mümkündür.

- `ax.legend(['1', '2'])`
- Burada:

  * `['1', '2']` — legend-də görünəcək etiketlərin adıdır və qrafikdəki xətlərə uyğun şəkildə göstərilir.

---

* **ax.legend()** – Qrafikdə legend-i göstərmək və xətlər ilə etiketləri əl ilə uyğunlaşdırmaq üçün istifadə olunur.

- `ax.legend((ax2, ax1), ['line2', 'line1'])`
- Burada:

  * `(ax2, ax1)` — legend-də göstəriləcək xətlərin obyektləri (plot-lar)
  * `['line2', 'line1']` — müvafiq olaraq xətlərə aid etiketlər
  * Bu yolla, legend-də xətlərin sırası və adları əl ilə idarə edilir.

---

* **ax.legend()** – Qrafikdə əfsanəni xüsusi riyazi ifadələr və əlavə üslub parametrləri ilə göstərmək üçün istifadə olunur.

- `ax.legend((ax2, ax1), [r'$f(x) = a \cdot b + c$', r'$f(x) = k \cdot x + b$'], facecolor='#aaa', framealpha=0.3)`
- Burada:

  * `(ax2, ax1)` —legend-də göstəriləcək xətlərin obyektləri
  * `[r'$f(x) = a \cdot b + c$', r'$f(x) = k \cdot x + b$']` — LaTeX formatında yazılmış riyazi düsturlar şəklində etiketlər
  * `facecolor='#aaa'` — legend-nin fon rəngini boz rəngdə təyin edir
  * `framealpha=0.3` — legend çərçivəsinin şəffaflıq səviyyəsini 0.3 (aşağı şəffaflıq) edir

---

* **Line2D()** – Matplotlib-də bir xətt obyektini (Line2D) yaratmaq üçün istifadə olunur, hansı ki, x və y koordinatlarından ibarət xətləri təsvir edir.

- `Line2D([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])`
- Burada:

  * `[0, 1, 2, 3, 4]` — x oxu üzrə nöqtələr
  * `[0, 1, 4, 9, 16]` — y oxu üzrə nöqtələr
  * Bu, x və y dəyərlərini birləşdirərək xət şəklində göstərir.
- **Qeyd:** Bu obyekt adi halda `matplotlib.lines` modulundan idxal edilir və qrafikdə istifadə üçün `Axes` obyektinə əlavə olunur.

---

* **ax.add\_line()** – Mövcud qrafikin (`Axes` obyektinin) üzərinə xüsusi `Line2D` xətt obyekti əlavə etmək üçün istifadə olunur.

- `ax.add_line(line)`
- Burada:

  * `line` — əvvəlcədən yaradılmış `Line2D` obyekti
  * Bu funksiya `Line2D` obyektini `ax` qrafikinə əlavə edir və qrafikdə göstərilməsini təmin edir.

---

* **Rectangle() və ax.add\_patch()** – Qrafikdə düzbucaqlı şəkil əlavə etmək üçün istifadə olunur.

- `rec = Rectangle((0, 0), 2.5, 0.5, facecolor='g')`
- `ax.add_patch(rec)`
- Burada:

  * `Rectangle((0, 0), 2.5, 0.5, facecolor='g')` — `(0, 0)` koordinatlarından başlayan, eni 2.5, hündürlüyü 0.5 olan yaşıl (`'g'`) rəngli düzbucaqlı obyekt yaradır
  * `ax.add_patch(rec)` — yaradılmış düzbucaqlını mövcud qrafikin üzərinə əlavə edir.
