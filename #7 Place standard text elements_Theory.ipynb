## Place standard text elements_Theory

---

* **plt.minorticks\_on()** – Həm **x**, həm də **y** oxları üçün **minor (kiçik) tick-ləri** (xırda aralıq nişanları) aktiv edir.

* **plt.grid(which='major', color='#444', linewidth=1)** – Yalnız **əsas (major)** grid xətlərini göstərir, **tünd rəngdə** (`#444`) və **1 vahid qalınlıqda**.

* **plt.grid(which='minor', color='#aaa', ls='-')** – Yalnız **kiçik (minor)** grid xətlərini göstərir, **açıq rəngdə** (`#aaa`) və **düz xətt tərzində** (`'-'`).

> Bu ayarlar əsas və kiçik grid xətlərini fərqli vizual stildə göstərməklə **qrafiki daha oxunaqlı və təmiz** edir.

---

* **fig.suptitle('Graphic')** – **Bütün qrafik fiqurunun yuxarısında** `'Graphic'` başlığını göstərir.

- `suptitle` – **“super title”**, yəni **bütün subplot-ları əhatə edən ümumi başlıq** deməkdir.
- Bu, `plt.title()`-dən fərqli olaraq **yalnız bir subplot üçün deyil, bütün `fig` üçün başlıq** verir.

> Əgər `fig, ax = plt.subplots()` istifadə etmisinizsə, bu metod `fig` üzərində işləyir.

---

* **ax.set\_xlabel('Ox', loc='left')** – **X oxunun adı** kimi `'Ox'` yazılır və **sol tərəfdə yerləşdirilir**.
* **ax.set\_ylabel('Oy', loc='center')** – **Y oxunun adı** kimi `'Oy'` yazılır və **şaquli olaraq ortada yerləşdirilir**.

> `loc` parametri ilə etiketi **sol**, **mərkəz** və ya **sağ** (x oxu üçün) / **yuxarı**, **mərkəz**, **aşağı** (y oxu üçün) kimi yerləşdirmək mümkündür:

* `loc='left'`, `loc='center'`, `loc='right'` (x üçün)
* `loc='top'`, `loc='center'`, `loc='bottom'` (y üçün)

---

**ax.text(0.1, 0.1, 'unknown project')** – Qrafik üzərində (0.1, 0.1) nöqtəsinə `'unknown project'` mətnini yerləşdirir.

---


**ax.annotate('annotate', xy=(0.6, 0.6), xytext=(0.8, 0.8), arrowprops={'facecolor':'gray', 'shrink':0.1})**


* **'annotate'** – Qrafikdə yazılacaq mətn (etiket).

* **xy = (0.6, 0.6)** – Oxun **ucu**nun göstərəcəyi nöqtə (yəni hansı yerə diqqət çəkiləcək).

* **xytext = (0.8, 0.8)** – Etiketin (mətnin) yerləşəcəyi nöqtə.

* **arrowprops = {...}** – Oxun vizual görünüşünü təyin edir:

  * **facecolor = 'gray'** – Oxun içi **boz rənglə** dolur.

  * **shrink = 0.1** – Oxun uzunluğunu **10% qısaldır**, yəni daha kompakt göstərir.

---

* **fig.suptitle()** – Şəkil (figure) üçün əsas başlıq əlavə etmək üçün istifadə olunur.
- `fig.suptitle('Graphic', fontsize=24, color='r')`
-  Burada `'Graphic'` başlığın mətni, `fontsize=24` şrift ölçüsü, `color='r'` isə başlığın rəngini qırmızı edir.

---

* **ax.text()** – Qrafikin müəyyən koordinatlarında mətn əlavə etmək üçün istifadə olunur.

- `ax.text(0.1, 0.1, 'unknown project', bbox={'boxstyle': 'darrow', 'facecolor': 'm'})`
- Burada:

  * `(0.1, 0.1)` — mətnin yerləşdiyi koordinatlar
  * `'unknown project'` — əlavə olunan mətn
  * `bbox` — mətn qutusunun üslubu və rəngi

    * `'boxstyle': 'darrow'` — ox şəklində qutu
    * `'facecolor': 'm'` — qutunun rəngi magenta
---


* **plt.figure()** – Yeni şəkil (figure) obyektini yaratmaq üçün istifadə olunur.

- `fig = plt.figure(figsize=(7, 4), facecolor='#eee')`
- Burada:

  * `figsize=(7, 4)` — şəkilin en və hündürlüyü düym (inch) ilə
  * `facecolor='#eee'` — şəkilin fon rəngi açıq boz (#eee hex kodu)

---

* **fig.set()** – Mövcud şəkilin (figure) xüsusiyyətlərini dəyişdirmək üçün istifadə olunur.

- `fig.set(facecolor='#eee')`
- Burada:

  * `facecolor='#eee'` — şəkilin fon rəngini açıq boz (`#eee` hex kodu) olaraq təyin edir.

---

* **ax.set\_facecolor()** – Qrafikin oxlarının (axes) fon rəngini təyin etmək üçün istifadə olunur.

- `ax.set_facecolor('#AAFFAA')`
- Burada:

  * `'#AAFFAA'` — açıq yaşıl rəngin hex kodudur və ox fonu bu rəngə boyanır.
