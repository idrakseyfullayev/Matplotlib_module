```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7,4))
ax  = fig.add_subplot()

x = np.arange(10)
np.random.seed(10)
y1 = np.random.randint(3, 20, len(x))
y2 = np.random.randint(3, 20, len(x))

w = 0.3

ax.bar(x - w/2, y1, width=w)
ax.bar(x + w/2, y2, width=w)

ax.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7,4))
ax  = fig.add_subplot()

vals = [10, 40, 23, 30 ,7]
labels = ['Toyoto', 'BMV', 'Lexus', 'Audi', "Lada"]

def exp():
    vals_ = []
    for i in vals:
        if i == max(vals):
            i = 0.2
        else:
            i = 0
            
        vals_.append(i)    
    return vals_ 

ax.pie(vals, labels = labels, autopct='%.2f', explode = exp(), shadow=True)

ax.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7,4))
ax  = fig.add_subplot()

vals = [10, 40, 23, 30 ,7]
labels = ['Toyoto', 'BMV', 'Lexus', 'Audi', "Lada"]

ax.pie(vals, labels = labels, autopct='%.2f', wedgeprops=dict(width=0.5), shadow=True)

ax.grid()
plt.show()
```
