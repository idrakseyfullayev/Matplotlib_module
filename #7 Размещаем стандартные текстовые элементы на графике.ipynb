```python
import numpy as np
import matplotlib.pyplot as plt

y = np.arange(0, 5, 1)
x = np.array([a*a for a in y])

y2 = [0, 2, 3, 4, 5, 7]
x2 = [i+1 for i in y2]

lines = plt.plot(x, y, x2, y2)

plt.minorticks_on()
plt.grid(which='major', color = '#444', linewidth = 1)
plt.grid(which='minor', color = '#aaa', ls = '-')

plt.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
plt.figtext= 0.05, 0.6, 'graphic'
fig.suptitle('Graphic')

ax.set_xlabel('Ox', loc = 'left')
ax.set_ylabel('Oy', loc = 'center')
ax.text(0.1, 0.1, 'unknown project')
ax.annotate('annotate', xy = (0.6, 0.6), xytext= (0.8, 0.8),
           arrowprops={'facecolor':'gray', 'shrink':0.1})
 
plt.show()
```
