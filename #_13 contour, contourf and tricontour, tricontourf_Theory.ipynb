## contour, contourf and tricontour, tricontourf_Theory

* **`ax[0].contour(x_grid, y_grid, z_grid)`** – `ax[0]` subplotunda kontur (izoxət) qrafiki çəkir.
- `x_grid`, `y_grid` – koordinat səthi üçün meshgrid-lərdir,
- `z_grid` – bu nöqtələrdəki funksiyanın dəyərləridir (hündürlük səthi).

---

* **`ax[1].contourf(x_grid, y_grid, z_grid)`** – `ax[1]` subplotunda dolu kontur (izoxət) qrafiki çəkir.
- `x_grid`, `y_grid` – koordinat səthi üçün meshgrid-lərdir,
- `z_grid` – bu nöqtələrdəki funksiyanın dəyərləridir (hündürlük səthi).

---

* **`c1 = ax[0].contour(x_grid, y_grid, z_grid)`** – `ax[0]` subplotunda kontur (izoxət) qrafiki çəkir və nəticəni `c1` dəyişəninə saxlayır.
-  **`ax[0].clabel(c1)`** – `c1` kontur xəttlərinin üzərinə səviyyə dəyərlərini əlavə edir.

---

* **`c1 = ax[0].contour(x_grid, y_grid, z_grid, 15)`** – `ax[0]` subplotunda 15 səviyyəli kontur (izoxət) qrafiki çəkir və `c1`-də saxlayır.
- **`c1.clabel()`** – `c1` kontur xəttlərinin üzərinə səviyyə dəyərlərini əlavə edir.

* Məsələn, səth funksiyasının dəyərləri 0-dan 100-ə qədər dəyişir.
* `contour` funksiyasında **15 səviyyə** təyin edəndə, bu 0-dan 100-ə qədər 15 bərabər aralıqla xətt çəkiləcək.
* Yəni 15 fərqli `z` dəyəri üzrə səth üzərində kontur xətləri olacaq.

---

* **`c1 = ax[0].contour(x_grid, y_grid, z_grid, [-0.1, 0, 0.2, 0.5])`** – `ax[0]` subplotunda kontur xətləri üçün səviyyələri `-0.1`, `0`, `0.2`, və `0.5` olaraq təyin edir və qrafiki çəkir.
- **`c1.clabel()`** – `c1` kontur xətlərinin üzərinə səviyyə dəyərlərini əlavə edir.

* Bu o deməkdir ki, yalnız `z = -0.1`, `z = 0`, `z = 0.2` və `z = 0.5` dəyərlərində kontur xətləri çəkiləcək.
* Digər bütün dəyərlər üçün kontur çəkilməyəcək.

---

* **`c1 = ax[0].contour(x_grid, y_grid, z_grid, cmap="plasma")`** – `ax[0]` subplotunda `"plasma"` rəng xəritəsi ilə kontur qrafiki çəkir və `c1` dəyişənində saxlayır.
- **`c1.clabel(colors='k')`** – `c1` kontur xətlərinin üzərinə qara (`'k'`) rəngdə səviyyə dəyərlərini əlavə edir.

---

* **`c1 = ax[0].contour(x_grid, y_grid, z_grid, cmap="plasma")`** – `ax[0]` subplotunda `"plasma"` rəng xəritəsi ilə kontur qrafiki çəkir və `c1`-də saxlayır.
- **`c1.clabel(colors='k', fmt='%.2f')`** – `c1` kontur xətlərinin üzərinə qara rəngdə və onluq dəqiqliklə (2 rəqəm) səviyyə dəyərlərini əlavə edir.

---

* **`c1 = ax.tricontour(x, y, z, cmap="plasma")`** – `ax` üzərində qeyri-müntəzəm paylanmış nöqtələrdən ibarət verilənlər üçün `"plasma"` rəng xəritəsi ilə triangulyasiya əsasında kontur qrafiki çəkir və `c1`-də saxlayır.
- **`c1.clabel(colors='k', fmt='%.2f')`** – `c1` kontur xətlərinin üzərinə qara rəngdə və onluq dəqiqliklə (2 rəqəm) səviyyə dəyərlərini əlavə edir.

---

- `tricontourf` funksiyası ilə yaradılan dolu kontur qrafiki üçün `clabel()` metodu mövcud deyil, ona görə də səviyyə dəyərlərini əlavə etmək mümkün deyil.

- **`c1 = ax.tricontourf(x, y, z, cmap="plasma")`** – `ax` üzərində qeyri-müntəzəm paylanmış nöqtələr üçün `"plasma"` rəng xəritəsi ilə dolu kontur qrafiki çəkir və `c1`-də saxlayır.

---


### **`contour` və `tricontour` arasındakı fərq:**

| Xüsusiyyət            | `contour`                                             | `tricontour`                                          |
| --------------------- | ----------------------------------------------------- | ----------------------------------------------------- |
| **Verilənlər tipi**   | Müntəzəm (grid) 2D nöqtələr (məsələn, `meshgrid` ilə) | Müntəzəm olmayan (scattered) 2D nöqtələr              |
| **Giriş verilənləri** | `X, Y, Z` — 2D massivlər (`shape=(m,n)`)              | `x, y, z` — 1D vektorlar (nöqtə koordinatları)        |
| **İstifadə sahəsi**   | Grid şəklində verilən səth üçün                       | Sərbəst paylanmış (scatter) nöqtələrdə səth üçün      |
| **Performans**        | Daha sürətli, amma yalnız grid üçün                   | Daha yavaşdır, amma qeyri-müntəzəm nöqtələrdə işləyir |



#### Sadə dildə:

- `contour` — **düzbucaqlı ızgara** (meshgrid) üzərində kontur çəkir.
- `tricontour` — **dağınıq, nizamsız nöqtələr** üzərində kontur çəkir.
