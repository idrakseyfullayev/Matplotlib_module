## The plot function_Theory



* **plt.plot()** – Verilmiş x və y massivlərini xətti qrafik şəklində çəkmək üçün istifadə olunur.

  `plt.plot(x, y)`

---

* **plt.show()** – Çəkilmiş qrafiki ekranda göstərmək üçün istifadə olunur.

  `plt.show()`

---

* **plt.grid()** – Qrafikdə köməkçi xətləri (şəbəkə xətləri) göstərmək üçün istifadə olunur.

  `plt.grid()`

---

* **plt.plot(x, y, linestyle='--')** – x və y dəyərlərini kəsik xətt (dashed line) ilə qrafikdə göstərmək üçün istifadə olunur.

  `plt.plot(x, y, '--')`

---

* **plt.plot(x, y, color='r')** – x və y dəyərlərini qrafikdə qırmızı rənglə göstərmək üçün istifadə olunur.

  `plt.plot(x, y, color='r')`

---

* **plt.plot(x, y, marker='+')** – x və y dəyərlərini qrafikdə “+” işarəsi ilə nöqtələyərək göstərmək üçün istifadə olunur.

  `plt.plot(x, y, marker='+')`

---

* **plt.plot(x, y, markerfacecolor='w')** – Qrafikdə nöqtələrin içini ağ (white) rənglə doldurmaq üçün istifadə olunur.

  `plt.plot(x, y, markerfacecolor='w')`

---

* **plt.plot(x, y, markersize=9)** – Qrafikdə nöqtələrin ölçüsünü 9 piksel olaraq təyin etmək üçün istifadə olunur.

  `plt.plot(x, y, markersize=9)`

---

* **plt.plot(x, y, linewidth=4)** – Qrafikdə xəttin qalınlığını 4 vahid olaraq təyin etmək üçün istifadə olunur.

  `plt.plot(x, y, linewidth=4)`

---

* **plt.setp()** – Qrafik obyektlərinin (məsələn, xətt, marker və s.) xüsusiyyətlərini dəyişdirmək üçün istifadə olunur.

  `plt.setp(line, color='r', linewidth=2)`

---

* **plt.plot(x, y)** – x və y dəyərlərini xətt qrafiki şəklində çəkmək üçün istifadə olunur.

- `plt.plot(x, y)`

* **plt.fill\_between(x, y)** – x oxu boyunca y və x oxunun aşağısı arasında sahəni rəngləmək üçün istifadə olunur.

- plt.fill_between(x, y)`

---

* **plt.fill\_between(x, y, y2=0.5)** – x oxu boyunca y ilə 0.5 səviyyəsi arasında olan sahəni rəngləmək üçün istifadə olunur.

  `plt.fill_between(x, y, y2=0.5)`

---

* **plt.fill\_between(x, y, where=y < 0, color='r', alpha=0.5)** – `y`-nin sıfırdan kiçik olduğu sahələri qırmızı, şəffaflıq səviyyəsi 0.5 ilə rəngləyir.

- `plt.fill_between(x, y, where=y < 0, color='r', alpha=0.5)`
  

* **plt.fill\_between(x, y, where=y > 0, color='g', alpha=0.5)** – `y`-nin sıfırdan böyük olduğu sahələri yaşıl, şəffaflıq səviyyəsi 0.5 ilə rəngləyir.

- `plt.fill_between(x, y, where=y > 0, color='g', alpha=0.5)`

---
