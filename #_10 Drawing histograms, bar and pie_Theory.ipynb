##  Drawing histograms, bar and pie_Theory

* **ax.hist()** – Verilənləri **histoqram** şəklində qruplaşdırıb göstərmək üçün istifadə olunur.

- `ax.hist(y)`
- Burada:

  * `y` — analiz ediləcək və sütunlara bölünəcək verilənlər
  * Bu funksiya `y` dəyərlərini **interval (bin)** qruplarına bölür və hər bir qrupda neçə dəyər olduğunu sütun hündürlüyü ilə göstərir
  * Əsasən **paylanmanı, sıxlıq bölgələrini** və ya **dağılım formasını** vizual analiz etmək üçün istifadə olunur.

---

* **ax.bar()** – Verilənləri **sütun qrafiki** (bar chart) şəklində göstərmək üçün istifadə olunur.

- `ax.bar(x, y)`
- Burada:

  * `x` — hər bir sütunun yerləşəcəyi mövqe (kateqoriyalar və ya indekslər)
  * `y` — hər bir sütunun hündürlüyü (dəyərlər)
  * Bu funksiya **kəmiyyətləri müqayisə etmək** və ya **kateqoriyalar üzrə paylanmanı göstərmək** üçün istifadə olunur.

---

Bu kod `bars` adlı list yaradır və aşağıdakıları edir:


`bars = [len(y[np.bitwise_and(y >= x[i], y < x[i+1])]) for i in range(len(x)-1)]`


### Qısa və sadə izah:

* `x[i], x[i+1]` → iki ardıcıl ədədi təyin edir: `[x[i], x[i+1])` aralığı.
* `y >= x[i]` və `y < x[i+1]` → `y` massivindəki hansı ədədlərin bu aralığa düşdüyünü tapır.
* `np.bitwise_and(...)` → bu iki şərti eyni anda tətbiq edir (və əməliyyatı).
* `y[...]` → həmin şərtə uyğun `y` dəyərlərini seçir.
* `len(...)` → həmin aralığa düşən `y` dəyərlərinin sayını tapır.
* `[ ... for i in range(len(x)-1)]` → bunu `x`-dəki hər aralıq üçün təkrar edir.

### Nəticə:

`bars` = `y` massivindəki dəyərlərin `x` aralıqları üzrə **paylanma sayları** (histogram kimi).

---

* **ax.barh()** – Verilənləri **üfüqi (horizontal) sütun qrafiki** kimi göstərmək üçün istifadə olunur.

- `ax.barh(x, y)`
- Burada:

  * `x` — sütunların yerləşəcəyi mövqe (adətən kateqoriyalar və ya indekslər)
  * `y` — hər bir sütunun **uzunluğu** (dəyəri)
  * Bu qrafik **bar()** funksiyasının üfüqi variantıdır və **məlumatları yan-yana müqayisə etmək** və ya **etiketlərin oxunmasını asanlaşdırmaq** üçün istifadə olunur.

---

* **ax.bar()** – Sütun qrafiki yaratmaq və onu vizual olaraq tənzimləmək üçün istifadə olunur.

- `ax.bar(x, y, width=0.5, linewidth=2, edgecolor='g', yerr=5, bottom=5)`
- Burada:

  * `x` — sütunların yerləşəcəyi mövqelər
  * `y` — hər sütunun hündürlüyü
  * `width=0.5` — sütunların eni (default 0.8-dir, burada isə daha nazikdir)
  * `linewidth=2` — sütunların sərhəd qalınlığı 2 piksel
  * `edgecolor='g'` — sərhəd rəngi yaşıl
  * `yerr=5` — hər sütun üçün 5 vahidlik **yuxarı-aşağı xəta çubuqları (error bars)** göstərilir
  * `bottom=5` — sütunlar 0-dan yox, **5-dən başlayır** (yəni bütün sütunlar aşağıya doğru 5 vahid yerdən “qalxır”)

Bu parametrlərlə sütunlar yaşıl sərhədli, 0.5 enində, 5 vahid aşağıdan başlayaraq çəkilir və üzərində 5 vahidlik xəta çubuqları göstərilir.

---

* **ax.bar()** – İki verilənlər qrupunu yanaşı sütunlar şəklində göstərmək üçün istifadə olunur.

- `w = 0.3`
- `ax.bar(x - w/2, y1, width=0.3)`
- `ax.bar(x + w/2, y2, width=0.3)`
- Burada:

  * `w = 0.3` — hər bir sütunun eni
  * `x - w/2` — `y1`-ə aid sütunlar bir az sola sürüşdürülür
  * `x + w/2` — `y2`-yə aid sütunlar bir az sağa sürüşdürülür
  * `width=0.3` — hər iki sütun qrupunun eni 0.3 olaraq təyin edilir

Bu üsulla eyni `x` mövqeyindəki `y1` və `y2` dəyərləri **yanaşı sütunlar (grouped bars)** şəklində göstərilir və kateqoriyalar üzrə müqayisə daha aydın görünür.

---

* **ax.pie()** – Verilənləri **dairəvi diaqram** (pie chart) şəklində göstərmək üçün istifadə olunur.

- `ax.pie(vals, labels=labels)`
- Burada:

  * `vals` — hər dilimin payını göstərən ədədi dəyərlər
  * `labels` — hər dilimə uyğun ad və ya kateqoriya etiketləri
  * Bu funksiya `vals` dəyərlərini **nisbətlərə çevirərək** dairəyə bölür və `labels` vasitəsilə hər hissəni tanıdır.

---

* **ax.pie()** – Dairəvi diaqramı (pie chart) daha detallı tənzimləmələrlə yaratmaq üçün istifadə olunur.

- `vals = [10, 40, 23, 30, 7]`
- `labels = ['Toyota', 'BMW', 'Lexus', 'Audi', 'Lada']`
- `exp = (0.1, 0.2, 0, 0, 0)`
- `ax.pie(vals, labels=labels, autopct='%.2f', explode=exp, shadow=True)`
  
- Burada:

  * `vals` — hər dilimin dəyəri
  * `labels` — dilimlərin adları
  * `autopct='%.2f'` — hər dilimin faizini iki onluq dəqiqliklə göstərir
  * `explode=exp` — dilimləri mərkəzindən müəyyən məsafədə ayrıdır (`0.1`, `0.2` və digərləri)
  * `shadow=True` — dairəvi diaqramın kölgəsini əlavə edir
- Nəticədə: `Toyota` və `BMW` dilimləri mərkəzdən biraz kənara çəkilərək vurğulanır, hər dilimin faizi göstərilir və diaqram kölgəli olur.

---

* **ax.pie()** – Dairəvi diaqramda ən böyük dilimi vurğulamaq üçün `explode` parametrindən istifadə edilərək dinamik şəkildə göstərmək üçün istifadə olunur.


- `vals = [10, 40, 23, 30, 7]`
- `labels = ['Toyoto', 'BMV', 'Lexus', 'Audi', 'Lada']`

- ```python
  def exp():
      vals_ = []
      for i in vals:
          if i == max(vals):
              i = 0.2
          else:
              i = 0
          vals_.append(i)
      return vals_
      
    ```    

- `ax.pie(vals, labels=labels, autopct='%.2f', explode=exp(), shadow=True)` 

- Burada:

  * `vals` — hər dilimin payını göstərən ədədi dəyərlər
  * `labels` — hər dilimə uyğun ad və ya kateqoriya etiketləri
  * `exp()` — maksimum dəyər üçün `explode=0.2` (dilim mərkəzdən uzaqlaşır), digərləri üçün isə `0`
  * `autopct='%.2f'` — faizlərin iki onluq dəqiqliklə göstərilməsi
  * `shadow=True` — diaqrama kölgə effekti əlavə edir
  * Bu funksiya `vals` dəyərlərini **nisbətlərə çevirərək** dairəyə bölür, ən böyük dilimi vurğulayır və `labels` vasitəsilə hissələri tanıdır.

---

* **ax.pie()** – Dairəvi diaqramı **halqa (donut) şəklində** göstərmək üçün istifadə olunur.

- `vals = [10, 40, 23, 30, 7]`
- `labels = ['Toyoto', 'BMV', 'Lexus', 'Audi', 'Lada']`
- `ax.pie(vals, labels=labels, autopct='%.2f', wedgeprops=dict(width=0.5), shadow=True)`
- Burada:

  * `vals` — dilimlərin dəyərləri
  * `labels` — dilimlərin adları
  * `autopct='%.2f'` — faizləri iki onluq dəqiqliklə göstərir
  * `wedgeprops=dict(width=0.5)` — dilimlərin enini 0.5 təyin edir və bu, diaqramı **halqa (donut) tipinə çevirir**
  * `shadow=True` — kölgə effekti əlavə edir
- Nəticədə dairəvi diaqramın mərkəzində dairəvi boşluq yaranır, yəni **halqa diaqramı** əldə edilir.
